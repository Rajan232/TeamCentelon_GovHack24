```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from datetime import datetime


# Load the data
def load_data(file_path):
    return pd.read_csv(file_path, low_memory=False)

# Function to print basic information about the dataframe
def basic_info(df, title):
    print(f"\n--- {title} ---")
    print(df.info())
    print("\nMissing values:")
    print(df.isnull().sum())
    print("\nDataset Description:")
    print(df.describe())

# Function to plot top categories in a column
def plot_top_categories(df, column, title, top_n=10):
    plt.figure(figsize=(12, 6))
    df[column].value_counts().nlargest(top_n).plot(kind='bar')
    plt.title(title)
    plt.ylabel('Count')
    plt.xlabel(column)
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    plt.show()

# Function to plot time-based trends
def plot_time_trend(df, date_column, value_column, title):
    plt.figure(figsize=(12, 6))
    df.groupby(date_column)[value_column].count().plot()
    plt.title(title)
    plt.xlabel('Date')
    plt.ylabel('Number of Accidents')
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()

# Function to categorize time into periods
def categorize_time(time):
    if ':' in time:
        hour = int(time.split(':')[0])
    else:
        hour = int(time[:2])
    if 0 <= hour < 6:
        return 'Night (00:00-05:59)'
    elif 6 <= hour < 12:
        return 'Morning (06:00-11:59)'
    elif 12 <= hour < 18:
        return 'Afternoon (12:00-17:59)'
    else:
        return 'Evening (18:00-23:59)'

# Light condition mapping
LIGHT_CONDITION_MAP = {
    1: 'Day',
    2: 'Dusk/Dawn',
    3: 'Dark Street lights on',
    4: 'Dark Street lights off',
    5: 'Dark No street lights',
    6: 'Dark Street lights unknown',
    9: 'Unknown'
}

# Severity mapping
SEVERITY_MAP = {
    1: 'Fatal accident',
    2: 'Serious injury accident',
    3: 'Other injury accident',
    4: 'Non injury accident'
}

# Speed zone mapping
def map_speed_zone(speed):
    if speed == 999:
        return 'Not known'
    elif speed == 777:
        return 'Other speed limit'
    elif speed == 888:
        return 'Camping grounds or off road'
    else:
        return f'{speed} km/hr'

def analyze_traffic_control_by_location(accident_df, vehicle_df, node_df):
    # Merge datasets
    merged_df = accident_df.merge(vehicle_df[['ACCIDENT_NO', 'TRAFFIC_CONTROL_DESC']], on='ACCIDENT_NO', how='left')
    merged_df = merged_df.merge(node_df[['ACCIDENT_NO', 'LGA_NAME', 'DEG_URBAN_NAME']], on='ACCIDENT_NO', how='left')

    # Analyze traffic control by LGA (Local Government Area)
    plt.figure(figsize=(15, 8))
    lga_control = merged_df.groupby('LGA_NAME')['TRAFFIC_CONTROL_DESC'].value_counts(normalize=True).unstack()
    lga_control_sorted = lga_control.sort_values('No control', ascending=False)
    lga_control_top20 = lga_control_sorted.head(20)
    lga_control_top20.plot(kind='bar', stacked=True)
    plt.title('Distribution of Traffic Control Types by Top 20 LGAs (Sorted by No Control)')
    plt.xlabel('Local Government Area')
    plt.ylabel('Proportion of Accidents')
    plt.legend(title='Traffic Control Type', bbox_to_anchor=(1.05, 1), loc='upper left')
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    plt.show()

    # Analyze traffic control by urban/rural classification
    plt.figure(figsize=(12, 6))
    urban_control = merged_df.groupby('DEG_URBAN_NAME')['TRAFFIC_CONTROL_DESC'].value_counts(normalize=True).unstack()
    urban_control.plot(kind='bar', stacked=True)
    plt.title('Distribution of Traffic Control Types by Urban/Rural Classification')
    plt.xlabel('Urban/Rural Classification')
    plt.ylabel('Proportion of Accidents')
    plt.legend(title='Traffic Control Type', bbox_to_anchor=(1.05, 1), loc='upper left')
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    plt.show()

    # Detailed analysis of 'No control' cases
    no_control_df = merged_df[merged_df['TRAFFIC_CONTROL_DESC'] == 'No control']

    plt.figure(figsize=(12, 6))
    sns.countplot(x='DEG_URBAN_NAME', data=no_control_df)
    plt.title('Distribution of No Control Accidents by Urban/Rural Classification')
    plt.xlabel('Urban/Rural Classification')
    plt.ylabel('Number of No Control Accidents')
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    plt.show()

    # Top 10 LGAs with highest proportion of 'No control' accidents
    lga_no_control_prop = no_control_df['LGA_NAME'].value_counts() / merged_df['LGA_NAME'].value_counts()
    plt.figure(figsize=(12, 6))
    lga_no_control_prop.sort_values(ascending=False).head(10).plot(kind='bar')
    plt.title('Top 10 LGAs with Highest Proportion of No Control Accidents')
    plt.xlabel('Local Government Area')
    plt.ylabel('Proportion of No Control Accidents')
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    plt.show()

# Analyze ACCIDENT dataset
def analyze_accident(df):
    basic_info(df, "ACCIDENT Dataset")

    # Flexible date parsing
    date_formats = ['%Y%m%d', '%Y-%m-%d', '%d-%m-%Y']
    for date_format in date_formats:
        try:
            df['ACCIDENT_DATE'] = pd.to_datetime(df['ACCIDENT_DATE'], format=date_format)
            print(f"Successfully parsed dates using format: {date_format}")
            break
        except ValueError:
            continue
    else:
        try:
            df['ACCIDENT_DATE'] = pd.to_datetime(df['ACCIDENT_DATE'], infer_datetime_format=True, dayfirst=True)
            print("Successfully parsed dates using pandas' flexible parsing")
        except ValueError:
            print("Failed to parse dates. Please check the date format in your data.")
            return

    # Analyze ACCIDENT_TIME
    df['TIME_CATEGORY'] = df['ACCIDENT_TIME'].apply(categorize_time)
    plot_top_categories(df, 'TIME_CATEGORY', 'Accidents by Time of Day')

    # Analyze LIGHT_CONDITION
    df['LIGHT_CONDITION_DESC'] = df['LIGHT_CONDITION'].map(LIGHT_CONDITION_MAP)
    plot_top_categories(df, 'LIGHT_CONDITION_DESC', 'Accidents by Light Condition')

    # Analyze SEVERITY
    df['SEVERITY_DESC'] = df['SEVERITY'].map(SEVERITY_MAP)
    plot_top_categories(df, 'SEVERITY_DESC', 'Distribution of Accident Severity')

    # Analyze SPEED_ZONE
    df['SPEED_ZONE_DESC'] = df['SPEED_ZONE'].apply(map_speed_zone)
    plot_top_categories(df, 'SPEED_ZONE_DESC', 'Accidents by Speed Zone')

    plot_top_categories(df, 'ACCIDENT_TYPE_DESC', 'Top 10 Accident Types')
    plot_top_categories(df, 'DAY_WEEK_DESC', 'Accidents by Day of Week')
    plot_time_trend(df, 'ACCIDENT_DATE', 'ACCIDENT_NO', 'Number of Accidents Over Time')

    return df

# Analyze ACCIDENT_EVENT dataset
def analyze_accident_event(df):
    basic_info(df, "ACCIDENT_EVENT Dataset")
    plot_top_categories(df, 'EVENT_TYPE_DESC', 'Top 10 Event Types')
    plot_top_categories(df, 'OBJECT_TYPE_DESC', 'Top 10 Object Types Involved')

# Analyze ACCIDENT_LOCATION dataset
def analyze_accident_location(df):
    basic_info(df, "ACCIDENT_LOCATION Dataset")
    plot_top_categories(df, 'ROAD_TYPE', 'Top 10 Road Types')
    plt.figure(figsize=(10, 6))
    plt.hist(df['DISTANCE_LOCATION'].dropna(), bins=30)
    plt.title('Distance from Intersection')
    plt.xlabel('Distance (m)')
    plt.ylabel('Frequency')
    plt.show()

# Analyze ATMOSPHERIC_COND dataset
def analyze_atmospheric_cond(df):
    basic_info(df, "ATMOSPHERIC_COND Dataset")
    plot_top_categories(df, 'ATMOSPH_COND_DESC', 'Distribution of Atmospheric Conditions')

# Analyze NODE dataset
def analyze_node(df):
    basic_info(df, "NODE Dataset")
    plot_top_categories(df, 'NODE_TYPE', 'Distribution of Node Types')
    plot_top_categories(df, 'LGA_NAME', 'Top 10 Local Government Areas')

# Analyze PERSON dataset
def analyze_person(df):
    basic_info(df, "PERSON Dataset")
    plot_top_categories(df, 'AGE_GROUP', 'Distribution of Age Groups')
    plot_top_categories(df, 'INJ_LEVEL_DESC', 'Distribution of Injury Levels')
    plot_top_categories(df, 'ROAD_USER_TYPE_DESC', 'Distribution of Road User Types')

# Analyze ROAD_SURFACE_COND dataset
def analyze_road_surface_cond(df):
    basic_info(df, "ROAD_SURFACE_COND Dataset")
    plot_top_categories(df, 'SURFACE_COND_DESC', 'Distribution of Road Surface Conditions')

# Analyze SUB_DCA dataset
def analyze_sub_dca(df):
    basic_info(df, "SUB_DCA Dataset")
    plot_top_categories(df, 'SUB_DCA_CODE_DESC', 'Top 10 Sub-DCA Codes')

# Analyze VEHICLE dataset
def analyze_vehicle(df):
    basic_info(df, "VEHICLE Dataset")
    plot_top_categories(df, 'VEHICLE_TYPE_DESC', 'Distribution of Vehicle Types')
    plot_top_categories(df, 'VEHICLE_MAKE', 'Top 10 Vehicle Makes')
    plot_top_categories(df, 'TRAFFIC_CONTROL_DESC', 'Distribution of Traffic Control Types')

    plt.figure(figsize=(12, 6))
    df['VEHICLE_YEAR_MANUF'] = pd.to_numeric(df['VEHICLE_YEAR_MANUF'], errors='coerce')
    df['VEHICLE_YEAR_MANUF'].dropna().hist(bins=30)
    plt.title('Distribution of Vehicle Year of Manufacture')
    plt.xlabel('Year')
    plt.ylabel('Count')
    plt.show()

# Function to analyze time-based patterns
def analyze_time_patterns(df):
    df['HOUR'] = df['ACCIDENT_TIME'].apply(lambda x: int(x[:2]))

    plt.figure(figsize=(12, 6))
    df['HOUR'].value_counts().sort_index().plot(kind='bar')
    plt.title('Distribution of Accidents by Hour')
    plt.xlabel('Hour of Day')
    plt.ylabel('Number of Accidents')
    plt.xticks(rotation=0)
    plt.tight_layout()
    plt.show()

# Function to analyze light condition vs severity
def analyze_light_condition_vs_severity(df):
    plt.figure(figsize=(12, 6))
    sns.countplot(x='LIGHT_CONDITION_DESC', hue='SEVERITY_DESC', data=df)
    plt.title('Accident Severity by Light Condition')
    plt.xlabel('Light Condition')
    plt.ylabel('Number of Accidents')
    plt.xticks(rotation=45, ha='right')
    plt.legend(title='Severity', bbox_to_anchor=(1.05, 1), loc='upper left')
    plt.tight_layout()
    plt.show()

def get_season(month):
    if month in [12, 1, 2]:
        return 'Summer'
    elif month in [3, 4, 5]:
        return 'Autumn'
    elif month in [6, 7, 8]:
        return 'Winter'
    else:
        return 'Spring'

# New function to analyze accidents by month and season
def analyze_accidents_by_time_period(df):
    df['Month'] = df['ACCIDENT_DATE'].dt.month
    df['Season'] = df['Month'].apply(get_season)

    plt.figure(figsize=(12, 6))
    df['Month'].value_counts().sort_index().plot(kind='bar')
    plt.title('Number of Accidents by Month')
    plt.xlabel('Month')
    plt.ylabel('Number of Accidents')
    plt.xticks(range(12), ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])
    plt.tight_layout()
    plt.show()

    plt.figure(figsize=(10, 6))
    df['Season'].value_counts().plot(kind='bar')
    plt.title('Number of Accidents by Season')
    plt.xlabel('Season')
    plt.ylabel('Number of Accidents')
    plt.tight_layout()
    plt.show()

    return df

# New function to analyze key factors by month
def analyze_key_factors_by_month(df):
    # Time of day analysis
    plt.figure(figsize=(12, 6))
    sns.countplot(x='Month', hue='TIME_CATEGORY', data=df)
    plt.title('Accidents by Time of Day and Month')
    plt.xlabel('Month')
    plt.ylabel('Number of Accidents')
    plt.xticks(range(12), ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])
    plt.legend(title='Time of Day')
    plt.tight_layout()
    plt.show()

    # Speed zone analysis
    high_speed = df[df['SPEED_ZONE'] >= 60]
    plt.figure(figsize=(12, 6))
    sns.countplot(x='Month', hue='SPEED_ZONE_DESC', data=high_speed)
    plt.title('Accidents in Speed Zones 60+ by Month')
    plt.xlabel('Month')
    plt.ylabel('Number of Accidents')
    plt.xticks(range(12), ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])
    plt.legend(title='Speed Zone', bbox_to_anchor=(1.05, 1), loc='upper left')
    plt.tight_layout()
    plt.show()


# Main analysis function
def main_analysis():
    # Load datasets
    accident_df = load_data('/content/sample_data/ACCIDENT.csv')
    accident_event_df = load_data('/content/sample_data/ACCIDENT_EVENT.csv')
    accident_location_df = load_data('/content/sample_data/ACCIDENT_LOCATION.csv')
    atmospheric_cond_df = load_data('/content/sample_data/ATMOSPHERIC_COND.csv')
    node_df = load_data('/content/sample_data/NODE.csv')
    person_df = load_data('/content/sample_data/PERSON.csv')
    road_surface_cond_df = load_data('/content/sample_data/ROAD_SURFACE_COND.csv')
    sub_dca_df = load_data('/content/sample_data/SUB_DCA.csv')
    vehicle_df = load_data('/content/sample_data/VEHICLE.csv')

    # Perform analysis on each dataset
    accident_df = analyze_accident(accident_df)
    analyze_accident_event(accident_event_df)
    analyze_accident_location(accident_location_df)
    analyze_atmospheric_cond(atmospheric_cond_df)
    analyze_node(node_df)
    analyze_person(person_df)
    analyze_road_surface_cond(road_surface_cond_df)
    analyze_sub_dca(sub_dca_df)
    analyze_vehicle(vehicle_df)

    # Analyze time-based patterns
    analyze_time_patterns(accident_df)

    # Analyze light condition vs severity
    analyze_light_condition_vs_severity(accident_df)

    accident_df = analyze_accidents_by_time_period(accident_df)
    analyze_key_factors_by_month(accident_df)

    analyze_traffic_control_by_location(accident_df, vehicle_df, node_df)
    # Perform joint analysis
    merged_df = accident_df.merge(accident_location_df, on='ACCIDENT_NO', how='left')
    merged_df = merged_df.merge(atmospheric_cond_df, on='ACCIDENT_NO', how='left')
    merged_df = merged_df.merge(road_surface_cond_df, on='ACCIDENT_NO', how='left')
    merged_df = merged_df.merge(vehicle_df[['ACCIDENT_NO', 'TRAFFIC_CONTROL_DESC']], on='ACCIDENT_NO', how='left')

    # Analyze relationships
    plt.figure(figsize=(10, 6))
    sns.boxplot(x='SEVERITY_DESC', y='SPEED_ZONE', data=merged_df)
    plt.title('Relationship between Severity and Speed Zone')
    plt.xticks(rotation=45, ha='right')
    plt.show()

    plt.figure(figsize=(12, 6))
    sns.countplot(x='SEVERITY_DESC', hue='ATMOSPH_COND_DESC', data=merged_df)
    plt.title('Severity by Atmospheric Condition')
    plt.xticks(rotation=45, ha='right')
    plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
    plt.tight_layout()
    plt.show()

    plt.figure(figsize=(12, 6))
    sns.countplot(x='SEVERITY_DESC', hue='SURFACE_COND_DESC', data=merged_df)
    plt.title('Severity by Road Surface Condition')
    plt.xticks(rotation=45, ha='right')
    plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
    plt.tight_layout()
    plt.show()

    plt.figure(figsize=(12, 6))
    sns.countplot(x='SEVERITY_DESC', hue='TRAFFIC_CONTROL_DESC', data=merged_df)
    plt.title('Severity by Traffic Control Type')
    plt.xticks(rotation=45, ha='right')
    plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
    plt.tight_layout()
    plt.show()

    # Analyze vehicle characteristics
    vehicle_accident_df = merged_df.merge(vehicle_df, on='ACCIDENT_NO', how='left')

    plt.figure(figsize=(12, 6))
    sns.countplot(x='SEVERITY_DESC', hue='VEHICLE_TYPE_DESC', data=vehicle_accident_df)
    plt.title('Severity by Vehicle Type')
    plt.xticks(rotation=45, ha='right')
    plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
    plt.tight_layout()
    plt.show()

    # Analyze the relationship between vehicle age and accident severity
    vehicle_accident_df['VEHICLE_AGE'] = pd.to_datetime(accident_df['ACCIDENT_DATE']).dt.year - pd.to_numeric(vehicle_df['VEHICLE_YEAR_MANUF'], errors='coerce')
    plt.figure(figsize=(10, 6))
    sns.boxplot(x='SEVERITY_DESC', y='VEHICLE_AGE', data=vehicle_accident_df)
    plt.title('Relationship between Vehicle Age and Accident Severity')
    plt.xticks(rotation=45, ha='right')
    plt.show()

# Run the analysis
if __name__ == "__main__":
    main_analysis()
```

**From our accident data analysis, we discovered some key patterns about when and where crashes happen most often. Here's what we found:**


Most car accidents occur during the busiest driving times - in the afternoon when people are heading home from work, and in the morning during the rush to work. These crashes frequently happen on city roads with speed limits of 60 km/h or higher. Drivers aged 30 to 39 are involved in accidents more than any other age group. Interestingly, most crashes take place when the road is dry, which many people think is safer for driving. This information shows us that even in what seems like good driving conditions, there's still a high risk of accidents. It's particularly important for drivers in their 30s to stay alert when driving in the city during busy times, even when the roads look safe and dry.

Roads with street lighting (likely urban areas) show high accident numbers.


##Vehicle and Road User Information:

After cars, station wagons and utility vehicles are commonly involved.
Motorcyclists, bicyclists, and pedestrians are significantly represented in accidents.
Vehicle-to-vehicle collisions are the most common type.


##Environmental Factors:

Over 75% of accidents occur in clear weather.
Rain is the second most common weather condition for accidents.
Autumn has slightly more accidents, followed closely by summer.
Most accidents happen on dry road surfaces.


##Road and Traffic Control:

Over 70% of accidents occur on roads and streets, as opposed to highways or freeways.
A striking majority of accidents happen in areas with no traffic control.
Areas with stop-go lights are the second most common for accidents with traffic control.


##Age and Severity Distribution:

The 30-39 age group is most frequently involved, closely followed by 40-49.
There's a gradual decline in accident involvement for ages over 50.
While most accidents result in injuries, the majority are not classified as severe.
Fatal accidents, though less frequent, occur across all light conditions.


##Urban vs. Rural Patterns:

Melbourne urban area accounts for the vast majority of accidents without traffic control.
Rural Victoria has the second-highest number of no-control accidents.
